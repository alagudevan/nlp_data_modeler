```python
file_name = "Pdf_image_test.pdf"
import platform
from tempfile import TemporaryDirectory
from pathlib import Path
 
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
from pytesseract import Output
```

```python
PDF_file = Path(file_name)
pdf_pages = convert_from_path(PDF_file, 500)

image_file_list = []
for page_enumeration, page in enumerate(pdf_pages, start=1):
    filename = f"page_{page_enumeration:03}.jpg" 
    page.save(filename, "JPEG")
    image_file_list.append(filename)

print(image_file_list)
```

```python
import cv2

def rotate_image_old(image_file):
    print(f"Rotate: {image_file}")
    img = cv2.imread(image_file)
    img_cw_180 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    img_cw_180.save(image_file, "JPEG")
    #cv2.imshow("Image rotated by 180 degree", img_cw_180)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    

def rotate_image(image_file, rotate_angle):
    print(f"Rotate: {image_file} to {360 - rotate_angle}")
    im = Image.open(image_file)
    im=im.rotate(360 - rotate_angle, expand=True)    
    im.save(image_file)
```

```python
#pytesseract.pytesseract.tesseract_cmd = r'C:\Users\USER\AppData\Local\Tesseract-OCR\tesseract.exe'
for image_file in image_file_list:
    results = pytesseract.image_to_osd(Image.open(image_file), output_type=Output.DICT)
    print(results)
    if int(results["rotate"]) > 0:
        rotate_image(image_file, int(results["rotate"]))
        results = pytesseract.image_to_osd(Image.open(image_file), output_type=Output.DICT)
        print(results)
    text = str(((pytesseract.image_to_string(Image.open(image_file)))))
    text = text.replace("-\n", "")
    print(text)
    
    
```
